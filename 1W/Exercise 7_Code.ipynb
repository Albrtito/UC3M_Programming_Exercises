```python
final_of_program = bool(False)
while final_of_program == False:
    #A is number1
    #B is number2
    number1 = float(input("A"))
    number2 = float(input("B"))

    if number1 == number2:
        print("Both numbers are equal.")
    if number2 == 0:
        print("The division is undefined")
    if number2 == 1 or number2 == -1:
        print("A will always be divisible by 1")
    else:

        # check if A>0
        if (number1 > 0 and number2 > 0) or (number1 < 0 and number2 < 0):
            #if A and B are < 0, then both minus signs cancel themselfs
            if number1 < 0 and number2 < 0 :
                number1 = -number1
                number2 = -number2
            # if B>0 repeat A-B while it is mayor than 0, or until it is equal or minor to 0
            while (number1 - number2) > -1:
                # if sometimes subtracting the operation is 0, A is divisible
                if number1 - number2 == 0:
                    print("A is divisible by B")
                    # return "A is divisible by B"
                    break
                # if none of these things happen for the A-B,
                # we store the A-B value and subtract A again in the next loop
                number1 -= number2


            # if the operation is minor to 0, it isn't
            if number1 - number2 < 0:
                # return "A is not divisible by B"
                print("A is not divisible by B")



        # check if A < 0
        if (number1 < 0 < number2) or (number1 > 0 > number2):
            if number1 > 0 > number2:
                number1 = -number1
                number2 = -number2
            while number1 + number2 < 1:
                if number1 + number2 == 0:
                    # return "A is divisible by B"
                    print("A is divisible by B")
                    break
                number1 += number2
                # breakpoint()

            if number1 + number2 > 0:
                # return "A is not divisible by B"
                print("A is not divisible by B")
        final_of_program = bool(input("Write values representing  True or False"))
```
